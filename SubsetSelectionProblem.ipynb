```python
from itertools import combinations

S = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
result = []

for subset in combinations(S, 5):
    if sum(subset) == 0:
        result.append(subset)

print("Subsets of size 5 with sum 0:")
for r in result:
    print(r)
```

```python
from itertools import combinations

S = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
result = []

for k in range(3, 7):  # from size 3 to 6
    for subset in combinations(S, k):
        if sum(subset) == 0:
            result.append(subset)

print("Subsets of size 3 to 6 with sum 0:")
for r in result:
    print(r)
```
