````python
from itertools import combinations

# Sample product list (can be read from file)
products = {
    'P1': 64,
    'P2': 88,
    'P3': 152,
    'P4': 93,
    'P5': 76,
    'P6': 120,
    'P7': 98,
    'P8': 130,
    'P9': 67,
    'P10': 95,
    'P11': 84,
    'P12': 110,
}

lower_limit = 490
upper_limit = 510

valid_combos = []

# Check combinations from size 2 to 5 (can be changed)
for r in range(2, 6):
    for combo in combinations(products.items(), r):
        ids = [item[0] for item in combo]
        total = sum(item[1] for item in combo)
        if lower_limit <= total <= upper_limit:
            valid_combos.append(ids)

# Print output
print("Valid combinations of products with total price in range:")
for combo in valid_combos:
    print(combo)
````
